# Load data from txt files

```python
import pandas as pd
```

```python
def txt_to_df(path):
    original_columns = pd.read_csv(path, sep="\t\t", nrows=0).columns
    cleaned_columns = [c.strip('"') for c in original_columns]
    df = pd.read_csv(
        path,
        sep="\t",
        skiprows=1,
        header=None,
        index_col=0,
        usecols=range(106),
    )
    df.columns = cleaned_columns
    return df.sort_index()

txt_to_df("../data/Breast/BREAST_Gene_Expression.txt")
```
